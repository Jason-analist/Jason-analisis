```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('../data/residuos-pet-santander.csv')

# Cálculos
df['kg_anual'] = df['kg_pet_mes'] * 12
df['ladrillos'] = df['kg_anual'] / 2
df['casas_vis'] = df['ladrillos'] // 2500
df['personas_beneficiadas'] = df['casas_vis'] * 4

# Gráfico 1: Casas posibles por municipio
plt.figure(figsize=(10,6))
plt.bar(df['municipio'], df['casas_vis'], color='#2E8B57')
plt.title('Casas VIS que se podrían construir con PET reciclado en Santander', fontsize=14)
plt.ylabel('Número de casas')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../img/dashboard.png')
plt.show()

# Insight clave
total_casas = df['casas_vis'].sum()
total_personas = df['personas_beneficiadas'].sum()
print(f"CONCLUSIÓN: Con el PET que bota Santander en 1 año, se le da techo a {total_personas} personas")
print(f"Total casas VIS posibles: {total_casas}")
```
